Q2:ITERATORS AND DECORATORS

Create a "Memory-Efficient Data Logger."
Decorator: Write a decorator @log_execution that prints the
name of the function being called and the time it took to run.
Generator: Write a generator function
stream_sensor_data(n) that yields n random temperature
readings one by one (simulating a real-time data stream).
Integration: Apply the decorator to a function that processes
the generated data.

```python
import time
import random

#decorator
def log_execution(func): #decorator definition
    def wrapper(*args, **kwargs):  #inner function that can take any no. of arguments and keyword arguments
        start = time.time() #records the time when function starts
        print(f"Calling {func.__name__} function")
        result = func(*args, **kwargs)
        end = time.time() #records the time when function ends
        print(f"Time taken to run the function is {end - start} seconds")
        return result
    return wrapper

#generator function
def stream_sensor_data(n):
    for i in range(n):
        temp = random.randint(20, 50)  #it generates a random number(temperature) between 20 and 50
        yield temp  #it generates the random temperature one by one and not all at once

#applying decorator to processor function
@log_execution  #processor = log_execution(processor)
def processor(temp): #it takes generator object as argument
    s = 0
    count = 0
    #for loop automatically performs next() to get another value
    for val in temp: #it takes one value at a time the generator generates and performs the sum and count to compute the average at last
        s = s + val
        count += 1
    if count == 0:
        return 0  #to handle empty data if n = 0
    avg = s / count
    return avg

temp_gen = stream_sensor_data(20)  #generator object
avg_temp = processor(temp_gen) #calls processor function but due to the decorator it actually calls the wrapper() function
print (f"Average of temperature is {avg_temp}") #prints the avg returned from return result statement of wrapper function
```

In the above Memory-Efficient Data Logger, there is the use of decorator 'log execution' to add log functionality to a 'processor' function that computes the average of the random temperatures that is generated by the generator stream_sensor(n).

The imports are for time calculation and random number generation.
Python reads the definitions at load time and does not run them until called.
As it sees @log_execution, it calls decorator function passing the processor function object as argument. Python reads the definitions of the wrapper function and then returns wrapper object from the decorator function.
Now, the processor function points the wrapper function. Then the program flow goes to the temp_gen a generator object which is passed to the processor function to compute average.Calling the processor function actually triggers the wrapper() because of the decorator. Inside the wrapper, the original processor function runs.
The wrapper function performs log operation as asked in the question and calls func(*args, **kwargs) which actually calls the processor function. The for loop inside processor requests the next value, and only then does the generator execute yield to produce one random temperature. It doesn't create all the numbers at once; it generates them one-by-one as the loop asks for them. This lazy evaluation is what makes it memory efficient. This function also computes average which is returned to avg_temp. It then gets printed.

Q 3: Object-Oriented Programming (OOP)

Task: Design a "Library Management System."
Encapsulation: Create a class Book with a "private" attribute
__is_borrowed. Provide methods to borrow and return.
Inheritance: Create a subclass DigitalBook that inherits from
Book but adds an attribute for file_format and overrides the
display method.

```python
class Book: #creating class Book
    def __init__(self,num, name): #constructor for class Book
        self.__is_borrowed = False #convention to name private attribute
        self.num = num
        self.name = name
    @property #getter to get the value of private attribute __is_borrowed
    def is_borrowed(self):
        return self.__is_borrowed
    @is_borrowed.setter #setter to set the value of private attribute
    def is_borrowed(self, status):
        if isinstance(status, bool):  #use of validation in setter
            self.__is_borrowed = status
        else:
            raise TypeError("Status must be True or False")

    def book_borrow(self): #borrow method
        if self.is_borrowed == True:
            print(f"Book {self.num} is already borrowed")
        else:
            print(f"You can borrow the book {self.num}")
            self.is_borrowed = True
    def book_ret(self): #return method
        if self.is_borrowed == False:
            print(f"Book {self.num} was not borrowed!")
            return
        print(f"Thank you for returning book {self.num}")
        self.is_borrowed = False
    def display(self): #display method that gets overridden by child class later
        print(f"Book no: {self.num}, Book name: {self.name}")

class DigitalBook(Book): #sub class DigitalBook that inherits Book class
    def __init__(self, num, name,file_format):
        super().__init__(num,name) #calling the constructor of parent class using super()
        self.file_format = file_format
    def display(self): #overriding the display method of parent class Book
        super().display() #performing the tasks of display() method of parent class
        print(f"File format : {self.file_format}")

b1 = Book('3001', 'Priya Sufi') #instantiating Book
print(b1.is_borrowed) #checking whether the book is borrowed or not
b1.book_borrow() #burrowing the book
print(b1.is_borrowed) #checking the value of is_borrowed after burrowing
b1.book_borrow() #trying to burrow already burrowed book
b1.book_ret() #returning the book
b1.book_borrow() #burrowing the same book again after returning
b1.display() #checking the details of the book

db1 = DigitalBook(200,"AI for all","PDF") #instantiating DigitalBook
print(f"Details of digital book {db1.num}")
db1.display() #use of overridden function



```

In the above Library management system, class named Book has been defined with its constructor performing the initialization of attributes from arguments and initializing the private attribute (naming convention using double underscore) __is_borrowed as False first.
We have set getter and setter to get the value of the private attribute and also change the value of the attribute safely using validation that the status is of boolean type.
The code contains borrow method and return method that handles the borrowing and returning of the book on the basis of is_borrowed boolean value. It checks whether the book is available to be borrowed or not and accordingly sets the value of __is_borrowed.
There is also a display method that displays the information of the book and is overridden by the child class DisplayBook.
DisplayBook is a derived class (subclass) that inherits the attributes and methods of Book. There is an extra attribute of it: file_format. In the constructor, we have also called the constructor of the Book class to initialize the attributes using super().
Also, there is method overriding of the method display which performs the task same as the parent class method (using super()) and also it performs extra task of displaying the file_format.
This is a small Library Management System using Object-Oriented Programming Concept in Python.
