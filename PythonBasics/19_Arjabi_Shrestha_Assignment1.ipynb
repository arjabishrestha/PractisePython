```python
import math #importing in different lines
import sys
def filter_num(threshold_num,numbers): #meaningful function name in snake_case
    result = [] #snakecase variable name
    for num in numbers: #meaningful variable names
        if num > threshold_num: #indentation has 4 spaces
            result.append(num * 10) #indentation has 4 spaces
    return result
threshold = 10 #meaningful variable name used
data_list = [1, 5, 12, 18, 2, 25] #variable following snake_case and spaces after comma in list
print("Filtered Data:" + str(filter_num(threshold,data_list)))

class StudentRecord: #class name follows PascalCase
    def __init__(self,name,s_id):
        self.name = name #lowercase attributes
        self.id = s_id
    def show(self): #method name in lowercase
        print(f"Student:{self.name}, ID:{self.id}")
s1=StudentRecord("John Doe","A001")
s1.show()
#space around operator is fixed in almost all lines
```
