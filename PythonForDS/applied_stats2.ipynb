```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "Hours":[2,4,6,8,10],
    "Marks":[50,65,80,82,97]
}
df=pd.DataFrame(data)
df
```

Skewness Calculation

```python
sk=df["Marks"].skew()
sk
```

Skewness Visualization using Histogram

```python
plt.hist(df["Marks"],bins=5)
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

Since from the above figure we can see the left tail is long, we say it is left skewed and also the calculation gives negative result.

Skewness Visualization using Boxplot

```python
plt.boxplot(x=df["Marks"])
plt.title("Marks Boxplot")
plt.show()
```

For a left skewed distribution, the upper whisker is shorter and lower whisker is longer.

Mean vs Median

```python
print(f"Mean:{df["Marks"].mean()}")
print(f"Median:{df["Marks"].median()}")
```

Mean is less than the median in left skewed. In right skewed, mean is greater than the median and in a symmetric distribution, mean equals to median.

Kurtosis

```python
k=df["Marks"].kurt()
k
```

Kurtosis can be visualized using histogram.

Normal like: Mesokurtic dataset

```python
data = [45,50,52,54,55,56,58,60,65]
df = pd.DataFrame({"values": data})
print(df["values"].kurt())
#histogram visualization
plt.hist(df["values"],bins=5)
plt.title("Mesokurtic kurtosis visualization")
plt.show()
plt.boxplot(df["values"])
plt.title("Boxplot")
plt.show()
```

Leptokurtic(heavy tails)

```python
data = [40, 42, 43, 45, 46, 48, 90]
df = pd.DataFrame({"values": data})
print(df["values"].kurt())
#histogram visualization
plt.hist(df["values"],bins=5)
plt.title("Leptokurtic kurtosis visualization")
plt.show()
#boxplot visualization for outliers
plt.boxplot(df["values"])
plt.title("Boxplot")
plt.show()
```

The histogram shows heavy tail and the boxplot shows that there are outliers which results in heavy tail.

Platykurtic

```python
data = [40,45,50,55,60,65,70,75]
df = pd.DataFrame({"values": data})
print(df["values"].kurt())
#histogram visualization
plt.hist(df["values"],bins=5)
plt.title("Platykurtic kurtosis visualization")
plt.show()
#boxplot visualization for outliers
plt.boxplot(df["values"])
plt.title("Boxplot")
plt.show()
```

Flat distribution and no outliers
