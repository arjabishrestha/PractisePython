Sampling and Random Variable

```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
data = {
    "student_id":range(1,1001),
    "marks":np.random.normal(70,10,1000),
    "study_hours":np.random.normal(5,2,1000)
}
df = pd.DataFrame(data)
df
```

```python
print("Understanding the population dataset")
print(f"Population mean of marks: {df["marks"].mean()}")
print(f"Population std of marks: {df["marks"].std()}")
print(f"Correlation between marks and study_hours {df["marks"].corr(df["study_hours"])}")
```

Taking a random sample

```python
sample=df.sample(n=50)
sample
```

Sample statistics calculation

```python
print("Understanding sample")
print(f"Sample mean of marks = {sample["marks"].mean()}")
print(f"Sample std of marks = {sample["marks"].std()}")
```

Comparing sample statistics and population parameters - they are not exactly same, but they are nearly equal as the sample size is large enough and also the sampling is done randomly.

Multiple Samples

```python
#5 samples of size 50
sample1=df.sample(n=50)
sample2=df.sample(n=50)
sample3=df.sample(n=50)
sample4=df.sample(n=50)
sample5=df.sample(n=50)
# mean of each samples
print(f"Mean of sample 1 is {sample1["marks"].mean()}")
print(f"Mean of sample 2 is {sample2["marks"].mean()}")
print(f"Mean of sample 3 is {sample3["marks"].mean()}")
print(f"Mean of sample 4 is {sample4["marks"].mean()}")
print(f"Mean of sample 5 is {sample5["marks"].mean()}")
```

We can see that the means differ slightly this shows sample variability.

Sampling distribution of mean

```python
sample_mean=[]
for i in range(200):
    s=df.sample(n=30)
    sample_mean.append(s["marks"].mean())
means=pd.Series(sample_mean)
print("Sample Mean distribution")
means.describe()
```

Sample of different sizes

```python
sampleA = df.sample(n=30)
sampleB = df.sample(n=50)
sampleC = df.sample(n=200)
A_mean=sampleA["marks"].mean()
B_mean=sampleB["marks"].mean()
C_mean=sampleC["marks"].mean()
print (f"Mean of sample size 30 is {A_mean}")
print (f"Mean of sample size 50 is {B_mean}")
print (f"Mean of sample size 200 is {C_mean}")
print (print (f"Mean of population is {df["marks"].mean()}"))
```

We can see that the mean of sample size 200 is nearly equal to the population mean. This shows that larger sample size gives better estimate.

Sampling with Replacement

```python
replacement = df.sample(n = 50, replace = True)
without = df.sample(n = 50, replace = False)
print(replacement["marks"].mean())
print(without["marks"].mean())
print(replacement["marks"].std())
print(without["marks"].std())
df["marks"].std()
```

Random Variables

```python
df["pass"] = df["marks"]>40
df
```

Probability of Passing

```python
prob_pass = (df[df["marks"]>50]["marks"].count())/df["marks"].count()
print(prob_pass)
# using mean
prob_pass=(df["marks"]>50).mean()
print(prob_pass)
```

Find probability of marks > 80

```python
prob_greater80 = (df["marks"]>80).mean()
print(prob_greater80)
```

Conditional Probability

P(pass/ study_hours > 6)

```python
study_hours_filtered = df[df["study_hours"] > 6]
P = (study_hours_filtered["marks"] > 50).mean()
print(P)
```
