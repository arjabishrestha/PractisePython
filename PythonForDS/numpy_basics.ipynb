Vectorized operations

```python
arr1=np.array([5,10,15,20])
print(arr1*2)
print(arr1+5)
print(arr1**2)
```

Aggregation functions

```python
mean=np.mean(arr1)
total=np.sum(arr1)
mini=np.min(arr1)
maxi=np.max(arr1)
print(f"Mean is {mean}")
print(f"Sum of values is {total}")
print(f"Minimum value is {mini}")
print(f"Maximum value is {maxi}")
```

```python
sd=np.std(arr1)
print(f"Standard deviation is {sd}")
```

Mini task: create array of student marks, find avg marks, highest marks, lowest marks, addd 5 bonus marks to all students

```python
marks=np.array([87,95,65,73,81,78])
avg_marks=np.sum(marks)/marks.size
highest=np.max(marks)
lowest=np.min(marks)
new_marks=marks+5
print(f"Average marks is {avg_marks}")
print(f"Highest marks is {highest}")
print(f"Lowest marks is {lowest}")
print(f"Adding 5 marks: {new_marks}")
```
