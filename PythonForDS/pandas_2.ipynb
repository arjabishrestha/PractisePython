```python
import pandas as pd
```

```python
df=pd.read_csv("students.csv")
df["marks"]=pd.to_numeric(df["marks"],errors="coerce")
df=df.dropna(subset=["marks","age"])
df
```

```python
df=df[df["marks"]>80]
df
```

```python
df=df[df["age"]>2]
df
```

```python
df=df[(df["marks"]>70) & (df["age"]>20)]
df
```

```python
df=df.sort_values("marks")
df
```

the above gives the values in ascending order. For descending order, add ascending=False.

```python
df=df.sort_values("marks",ascending=False)
df
```

Add new column using vectorization.

```python
df["bonus marks"]=df["marks"]+5
df
```

Basic aggregation

```python
print(f"Mean= {df["marks"].mean()}")
print(f"Max= {df["marks"].max()}")
print(f"Min= {df["marks"].min()}")
```

Adding new column GRADE using vectorization (pd.cut())

```python
df["grade"]=pd.cut(
    df["marks"],
    bins=[0,60,80,100],
    labels=['C','B','A'],
    right=False
)
df
```

Adding new column pass or fail using vectorization

```python
df["Result"]=pd.cut(
    df["marks"],
    bins=[0,80,100],
    labels=["Fail","Pass"],
    right=False
)
df
```

Analysis

Which grade has highest average marks

```python
grade_wise_avg=df.groupby("grade")["marks"].mean()
print(grade_wise_avg)
print(grade_wise_avg.idxmax())
```

How many students are in each grade

```python
count_by_grade=df["grade"].value_counts()
count_by_grade
```

Who has the highest marks?

```python
top=df.loc[df["marks"].idxmax()]
top
```

Students scoring above average marks

```python
overall_mean=df["marks"].mean()
print(overall_mean)
print(df[df["marks"]>overall_mean])
```
