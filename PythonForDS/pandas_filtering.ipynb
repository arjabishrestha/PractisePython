```python
import pandas as pd
```

```python
df=pd.read_csv("sales.csv")
df["price"]=pd.to_numeric(df["price"],errors="coerce")
df["quantity"]=pd.to_numeric(df["quantity"],errors="coerce")
df=df.dropna(subset=["price","quantity"])
df["amount"]=df["price"]*df["quantity"]
df.head()
```

Task 1: Details of products where price>100

```python
filtered_df=df[df["price"]>100]
filtered_df["product"]
```

Task 2: To show rows where quantity>=5

```python
filtered=df[df["quantity"]>=5]
filtered
```

Task 3: To show rows where price>1000 and quantity>2

```python
filtered1=df[(df["price"]>1000)&(df["quantity"]>2)]
filtered1
```

Sorting dataset by amount (highest first)

```python
sorted_df=df.sort_values(by="amount",ascending=False)
sorted_df
```

Create column: Price label

```python
def price_label(price):
    if price>5000:
        return "High"
    elif price>2000 and price<=5000:
        return "Medium"
    else:
        return "Low"
df["Price Label"]=df["price"].apply(price_label)
df
```

Dataset Exploration

Number of products per category

```python
numpercat=df["category"].value_counts()
numpercat
```

Unique Categories

```python
unique_cat=df["category"].unique()
unique_cat
```

```python
print(f"Total categories is {df["category"].nunique()}")
```

Mini Analysis

1. Which product generated the highest amount?

```python
highest_amt=df["amount"].max()
high_amt_prod=df[df["amount"]==highest_amt]
print(high_amt_prod)
high_amt_prod
```

```python
high_amt_prod=df.loc[df["amount"].idxmax(),["product","category"]]
print(f"{high_amt_prod} generated the highest amount")
```

2. Which category has most products?

```python
cat_wise=df.groupby("category")["product"].nunique()
print(f"{cat_wise.idxmax()} category has most products")
```

3. Average price per category

```python
avg_price_cat=df.groupby("category",as_index=False)["price"].mean()
avg_price_cat=avg_price_cat.rename(columns={"price":"avg price"})
avg_price_cat
```
