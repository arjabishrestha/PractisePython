```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales_data=pd.read_csv("sales.csv")
sales_data["price"]=pd.to_numeric(sales_data["price"],errors="coerce")
sales_data["quantity"]=pd.to_numeric(sales_data["quantity"],errors="coerce")
sales_data=sales_data.dropna(subset=["price","quantity"])
sales_data
```

```python
sales_data["amount"]=sales_data["price"]*sales_data["quantity"]
sales_data
```

bar chart - to show total sales per category

```python
total_cat=sales_data.groupby("category")["amount"].sum()
total_cat
```

```python
plt.bar(total_cat.index,total_cat.values)
plt.title("Total sales per category")
plt.xlabel("Category")
plt.ylabel("Total sales")
plt.show()
```

The above bar graph shows the total sales per category. From the graph, we can see that the Accessories dominates the total sales and the difference in the sales of Accessories and Stationery is medium. This shows that Accessories are sold more than stationery.

Histogram- distribution of price

```python
plt.hist(sales_data["price"], bins=5)
plt.title("Price distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

Histogram is for understanding spread

Scatter Plot- Price vs Qty

```python
plt.scatter(sales_data["price"],sales_data["quantity"])
plt.title("Price Vs Quantity")
plt.xlabel("Price")
plt.ylabel("Quantity")
plt.show()
```

From the above scatter plot, we see that as the price increases, the quantity decreases.

Boxplot using seaborn

```python
sns.boxplot(x=sales_data["price"])
plt.title("Price boxplot")
plt.show
```

Histogram for quantity

```python
plt.hist(sales_data["quantity"], bins=10)
plt.title("Histogram: Quantity")
plt.xlabel("Quantity")
plt.ylabel("Frequency")
plt.show()
```

Barchart - number of products per category

```python
products_per_cat=sales_data.groupby("category")["product"].count()
plt.bar(products_per_cat.index,products_per_cat.values)
plt.title("Number of products per category")
plt.xlabel("Category")
plt.ylabel("Number of products")
plt.show()
```
